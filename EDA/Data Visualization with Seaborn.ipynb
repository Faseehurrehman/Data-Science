## Data

```python
import pandas as pd
import numpy as np
uniform_data = np.random.rand(10, 12)
data = pd.DataFrame({'x':np.arange(1,101), 'y':np.random.normal(0,4,100)})
```

```python
# Built-in Datasets in Seaborn
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

## Figure Aesthetics

```python
f, ax = plt.subplots(figsize=(5,6)) # Create a figure and one subplot
```

```python
# Seaborn Styles
sns.set() # Reset the seaborn default
sns.set_style("whitegrid") # Set the matplotlib parameters
sns.set_style("ticks",  {"xtick.major.size":8, "ytick.major.size":8}) # Set the matplotlib parameters
sns.axes_style("whitegrid") # Return a dict of params or use with to temporarily set the style
```

```python
# Context Function
sns.set_context("talk")
```

```python
# Color Palette
sns.set_palette("husl",3) # Define the color palette
sns.color_palette("husl") # Use with with to temporarily set palette
flatui = ["#9b59b6","#3498db","#95a5a6","#e74c3c","#34495e","#2ecc71"]
sns.set_palette(flatui) # Set your own color palette
```

## Plotting With Seaborn

#### Axis Grids

```python

g = sns.FacetGrid(titanic, col="survived", row="sex") # Subplot grid for plotting conditional relationships
```

```python
 g = g.map(plt.hist,"age")
```

```python
sns.factorplot(x="pclass", y="survived", hue="sex", data=titanic) #  Draw a categorical plot onto a Facetgrid
```

```python
 sns.lmplot(x="sepal_width", y="sepal_length", hue="species", data=iris) # Plot data and regression model fits across a FacetGrid
```

```python
h = sns.PairGrid(iris) # Subplot grid for plotting pairwise relationships
```

```python
h = h.map(plt.scatter)
```

```python
sns.pairplot(iris) # Plot pairwise bivariate distributions
```

```python
i = sns.JointGrid(x="x", y="y", data=data) # Grid for bivariate plot with marginal univariate plots
```

```python
i = i.plot(sns.regplot, sns.distplot)
```

```python
sns.jointplot("sepal_length", "sepal_width", data=iris, kind='kde') # Plot bivariate distribution
```

#### Regression Plots

```python
sns.regplot(x="sepal_width", y="sepal_length", data=iris, ax=ax) # Plot data and a linear regression model fit
```

#### Distribution Plots

```python
plot = sns.distplot(data.y, kde=False, color="b") # Plot univariate distribution
```

#### Matrix Plots

```python
sns.heatmap(uniform_data,vmin=0,vmax=1)
```

#### Categorical Plots

```python
# Scatterplot
sns.stripplot(x="species", y="petal_length", data=iris) # Scatterplot with one categorical variable
```

```python
sns.swarmplot(x="species", y="petal_length", data=iris) # Categorical scatterplot with non-overlapping points
```

```python
# Bar Chart
sns.barplot(x="sex", y="survived", hue="class", data=titanic) # Show point estimates and confidence intervals with scatterplot glyphs
```

```python
# Count Plot
sns.countplot(x="deck", data=titanic, palette="Greens_d") # Show count of observations
```

```python
# Point Plot
sns.pointplot(x="class", y="survived", hue="sex", data=titanic, palette={"male":"g", "female":"m"}, markers=["^","o"], linestyles=["-","--"]) # Show point estimates and confidence intervals as rectangular bars
```

```python
# Boxplot
sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic) #  Boxplot
```

```python
sns.boxplot(data=iris,orient="h") # Boxplot with wide-form data
```

```python
# Violinplot
sns.violinplot(x="age", y="sex", hue="survived", data=titanic) #  Violin plot
```

## Further Customizations

#### Axisgrid Objects

```python
g.despine(left=True) # Remove left spine
g.set_ylabels("Survived") # Set the labels of the y-axis
g.set_xticklabels(rotation=45) # Set the tick labels for x
g.set_axis_labels("Survived", "Sex") # Set the axis labels
h.set(xlim=(0,5), ylim=(0,5),  xticks=[0,2.5,5], yticks=[0,2.5,5]) # Set the limit and ticks of the x-and y-axis
```

#### Plot

```python
plt.title("A Title") # Add plot title
plt.ylabel("Survived") # Adjust the label of the y-axis
plt.xlabel("Sex") # Adjust the label of the x-axis
plt.ylim(0,100) # Adjust the limits of the y-axis
plt.xlim(0,10) # Adjust the limits of the x-axis
plt.setp(ax,yticks=[0,5]) # Adjust a plot property
plt.tight_layout() # Adjust subplot params
```
