## Single Line Plots

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
plt.plot(x)
plt.show()
```

```python
plt.plot(x,[y**2 for y in x]) # OR plt.plot(x,x**2)
plt.show()
```

```python
x = [1,4,2,5,3,6]
plt.plot(x)
```

## Multiline Plots

```python
plt.figure(figsize=(15,10))
x = np.arange(10)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.plot(x,2*x)
plt.plot(x,x+2)

plt.show()
```

```python
# Doing the above process in a single line
plt.figure(figsize=(15,10))
x = np.arange(10)
plt.plot(x,x**2,  x,x**3,  x,2*x,  x,x+2)

plt.show()
```

```python
x = np.array([[1,2,3,8],[4,5,6,8]])
plt.plot(x)
plt.show()
```

```python
x = np.random.randn(2,10)
print(x)
plt.plot(x[0],x[1])
plt.show()
```

```python
x = np.arange(10)
plt.plot(x,x**2,  x,x**3,  x,2*x,  x,x+2)
plt.grid(True)

plt.show()
```

```python
plt.plot(x,x**2,  x,x**3,  x,2*x,  x,x+2)
plt.grid(True)
plt.axis([0,2, 0,5]) # Changing range of axes   OR we can differently as: plt.xlim([0,2])   plt.ylim([0,5])
 
plt.show()
```

```python
plt.plot(x,x**2,  x,x**3,  x,2*x,  x,x+2)
plt.grid(True)
plt.xlim([0,2])
plt.ylim([0,5])
plt.xlabel('x = np.arange(10)')
plt.ylabel('y=f(x)')
plt.title('Multiline Plot')

plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,x**2, label = 'x**2')
plt.plot(x,x**3, label = 'x**3')
plt.plot(x,2*x, label = '2*x')
plt.plot(x,x+2, label = 'x+2')
plt.grid(True)
plt.legend()
plt.xlim([0,2])
plt.ylim([0,5])
plt.xlabel('x = np.arange(10)',fontsize=12,color='green')
plt.ylabel('y=f(x)',fontsize=12,color='green')
plt.title('Multiline Plot',fontsize=15,color='red')
plt.savefig('Multiline plot.png')

plt.show()
```
