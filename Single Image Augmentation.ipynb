```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('rabbit.30.jpg',target_size = (200,200))
```

```python
img
```

```python
type(img)
```

```python
img = image.img_to_array(img)
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3) # 1 means number of images we are going to process
```

```python
input_batch.shape
```

```python
datagen = ImageDataGenerator(
    rotation_range = 40,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range= 0.2,
    height_shift_range= 0.2,
    fill_mode='constant'
)
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='train'): # we use flow if we have single image, we will have one batch size in output
  i = i+1
  if i ==10:
    break
```
