```python
iqr = percentile75 - percentile25
iqr
```

### Model Building

```python
X=df13.drop('price',axis='columns') #To drop the dependent column
y = df13['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)    #20% data is for evaluation of Model performance
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
def rmse(y_test, y_pred):
  return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
lr_rmse = rmse(y_test, lr.predict(X_test))
lr_score=lr.score(X_test,y_test)
lr_score, lr_rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
rfr_rmse = rmse(y_test, rfr.predict(X_test))
rfr_score=lr.score(X_test,y_test)
rfr_score, rfr_rmse
```
