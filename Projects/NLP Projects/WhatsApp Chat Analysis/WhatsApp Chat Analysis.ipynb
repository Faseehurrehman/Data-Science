```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
f = open('WhatsApp Chat with VAILI Team.txt','r',encoding='utf-8')
data = f.read()
```

```python
pattern = '\d{1,2}\/\d{1,2}\/\d{2,4}\,\s\d{1,2}\:\d{2}\s\DM\s\-\s'
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_messages':messages,'message_dates':dates})
df
```

```python
users = []
messages = []

for i in df['user_messages']:
    entry = re.split('([\w\W]+?):\s', i)
    if entry[1:]: # User Names
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])


df['user'] = users
df['message'] = messages
        
```

```python
df.drop(columns=['user_messages'], inplace=True)
```

```python
df['message_dates'] = df['message_dates'].str.replace('-','')
```

```python
df['message_dates'] = df['message_dates'].str.strip()
```

```python
df['message_dates']
```

```python
df['message_dates'] = pd.to_datetime(df['message_dates'])
```

```python
df['message_dates']
```

```python
# Feature Extraction

df['dates'] = df['message_dates'].dt.date
df['days'] = df['message_dates'].dt.day
df['years'] = df['message_dates'].dt.year
df['months'] = df['message_dates'].dt.month
df['day_names'] = df['message_dates'].dt.day_name()
df['month_names'] = df['message_dates'].dt.month_name()
df['hours'] = df['message_dates'].dt.hour
df['minutes'] = df['message_dates'].dt.minute
```

```python
# Removal of Stop Words

f = open('stop_hinglish.txt','r',encoding = 'utf-8')
stop_words = f.read()
print(stop_words)
```

```python
words = []

for i in df['message']:
    for word in i.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
# Messages year and month-wise
timeline = df.groupby(['years','month_names','months']).count()['message'].reset_index()
```

```python
timeline
```

```python
# Merging
time = []
for i in range(timeline.shape[0]):
    time.append(timeline['month_names'][i] + '-' + str(timeline['years'][i]))
```

```python
time
```

```python
timeline['time']=time
timeline
```

```python
plt.plot(timeline['time'],timeline['message'])
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# Heatmap - On which time user is most active

period = []
for hour in df[['day_names', 'hours']]['hours']:
        if hour == 23:
            period.append(str(hour) + "-" + str('00'))
        elif hour == 0:
            period.append(str('00') + "-" + str(hour + 1))
        else:
            period.append(str(hour) + "-" + str(hour + 1))

df['period'] = period
```

```python
df.head(2)
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.pivot_table(index = 'day_names', columns = 'period', values = 'message', aggfunc = 'count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
