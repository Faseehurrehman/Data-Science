```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://qoshe.com/pk/en').text
```

```python
soup=BeautifulSoup(webpage,'html.parser')
```

```python
data = soup.find_all('div',attrs={'class':'box'})
```

```python
data
```

```python
images = []
paragraph = []
author_names = []
titles = []

for i in data:
    name = i.h1.text if i.h1 else ''
    titles.append(name)
    
    img = i.find('img')
    if img:
        images.append(img['src'])
    else:
        images.append(' ')
 
    
    author = i.find('div','info').h2.text if i.find('div','info') else ''
    author_names.append(author)
    
  
    
    para = i.p if i.p else ''
    paragraph.append(para)
```

```python
df = pd.DataFrame({'Titles':titles,'Author Names':author_names,'Paragraph':paragraph,'Images':images})
```
