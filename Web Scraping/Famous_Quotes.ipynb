```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import numpy as np
from time import sleep
from random import randint
```

```python
pages = np.arange(1,1000,20)
```

```python
titles = []
author_names = []
likes = []
links = []
images = []
```

```python
pages = np.arange(1,600,30)
```

```python
for page in pages:
    page = requests.get('https://www.goodreads.com/quotes?page='+str(page))
    soup = BeautifulSoup(page.text, 'html.parser')
    data = soup.find_all('div',attrs={'class':'quote'})
    sleep(randint(2,8))
    for i in data:
        text = i.find('div', class_='quoteText').text.replace('\n','')
        titles.append(text)
    
        author = i.find('div', class_='quoteText').find('span', class_='authorOrTitle').text.replace('\n','')
        author_names.append(author)
    
        like = i.find('div', class_ = 'right').find('a', class_='smallText').text.replace('likes','').replace(' ','')
        likes.append(like)
        
        anchor = i.find('a')
        img = i.find('img')
        links.append(anchor['href'])
        
        if img:
          images.append(img['src'])
        else:
          images.append(' ')
        
```

```python
len(titles)
```

```python
df = pd.DataFrame({'Quotes':titles,'Author Names':author_names,'Links':links,'Images':images,'Likes':likes})
```
