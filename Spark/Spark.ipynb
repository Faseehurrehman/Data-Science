```python
# For words count
string = "He then joined the Christian Mission High School where his parents thought his restless mind could be focused. Karachi proved more prosperous for young Jinnah than Bombay had been. His father's business had prospered so much by this time that he had his own stables and carriages. Jinnahbhai Poonja's firm was closely associated with the leading British managing agency in Karachi, Douglas Graham and Company. Sir Frederick Leigh Croft, the general manager of the company, had a great influence over young Jinnah, which possibly lasted his entire life"
string.split(" ")
```

```python
words_mapped = file.map(lambda x: x.split(" "))
words_mapped.collect()
```

```python
words_mapped.count()
```

```python
flat_words_mapped = file.flatMap(lambda x: x.split(" "))
flat_words_mapped.count() # Number of words
```

```python
filtered_words = flat_words_mapped.filter(lambda x: len(x)>5)
filtered_words.count()
```

```python
pairs = flat_words_mapped.flatMap(lambda x:(x,1))
pairs.collect()
```

```python
pairs = flat_words_mapped.map(lambda x: (x,1))
pairs.collect()
```

```python
grouped_pairs = pairs.groupByKey()
```

```python
grouped_pairs.collect()
```

```python
for (key, value) in grouped_pairs.collect():
    print(key, len(list(value)))  
```

```python
pairs = flat_words_mapped.map(lambda x:(x[0:1],x))
```

```python
pairs.collect()
```

```python
grouped_pairs = pairs.groupByKey()
grouped_pairs.collect()
```

```python
for (key,value) in grouped_pairs.collect():
    print(key,len(list(value)))
```
