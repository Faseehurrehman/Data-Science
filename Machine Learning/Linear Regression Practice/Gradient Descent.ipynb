# To derive best fit line using m and b values

<b> We have been given x and y vectors. So we have to find correct values of m and b for best fit line

```python
import numpy as np
```

```python
def gradient_descent(x,y):
    # Initiate with some values
    m_current=b_current=0   # Initiate with some values and then by taking small steps we will try to reach global minima
    iterations=10000     # Initiate with some values and then fine tune them
    learning_rate=0.001  # Initiate with some values and then fine tune them
    n=len(x)             # length of data points
    
    for i in range(iterations):
        y_predicted=m_current * x + b_current
        cost=(1/n)*sum(([val**2 for val in (y-y_predicted)]))
        md=-(2/n)*sum(x*(y-y_predicted))  # partial derivate of m
        bd=-(2/n)*sum(y_predicted)      # partial derivative of b
        m_current=m_current - learning_rate * md
        b_current=b_current - learning_rate * bd
        print("m {}, b {}, cost {}, iteration {}".format(m_current,b_current,cost,i))
        
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```

<b> It shows that global minimum reached at iteration 157. Thus, here values of m and b will be best for best fit line
